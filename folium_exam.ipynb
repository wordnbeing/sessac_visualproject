# Markers

```python
m = folium.Map(
    location=[37.53, 126.97],
    tiles='Stamen Terrain',
    zoom_start=13
)

tooltip = 'Click me!'

folium.Marker(
    [37.55, 126.97],
    popup = '<b>Seoul<b>',
    tooltip = tooltip
).add_to(m)

folium.Marker(
    [37.52, 126.96],
    popup = '<b>YongSan<b>',
    tooltip = tooltip
).add_to(m)

m
```

# icons

```python
m = folium.Map(
    location=[37.53, 126.97],
    tiles='Stamen Terrain',
    zoom_start=13
)

folium.Marker(
    [37.55, 126.97],
    popup = '<b>Seoul<b>',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)

folium.Marker(
    [37.52, 126.96],
    popup = '<b>YongSan<b>',
    icon=folium.Icon(color='green', icon='bookmark')
).add_to(m)

m
```
